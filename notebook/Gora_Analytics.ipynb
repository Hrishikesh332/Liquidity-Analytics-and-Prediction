# **Total Borrow vs Average Borrow Amount**

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
from plotly.offline import plot


fig = px.scatter(df, x='total_borrow', y='avg_borrow_amount', hover_data=['address'])
fig.update_layout(
    title='Total Borrow vs Average Borrow Amount',
    plot_bgcolor='black',
    paper_bgcolor='black',
    font_color='green'
)


annotation_text = [f'Total Borrow: {x}<br>Avg Borrow: {y}' for x, y in zip(fig.data[0].x, fig.data[0].y)]
fig.update_traces(
    text=annotation_text,
    textposition='top center',
    hoverinfo='none'
)
fig.show()
```

```python
import pandas as pd
import plotly.express as px
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output


app = dash.Dash(__name__)


app.layout = html.Div([
    html.H1('Analytics Dashboard'),
    dcc.Tabs([
        dcc.Tab(label='Overview', children=[
            html.Div([
                dcc.Graph(
                    figure=px.scatter(df, x='total_borrow', y='total_repay', color='token_borrow_mode',
                                      hover_data=['address', 'avg_borrow_amount', 'avg_repay_amount'])
                ),
                dcc.Graph(
                    figure=px.histogram(df, x='days_since_first_borrow', color='liquidated')
                )
            ], style={'display': 'flex'}),
            html.Div([
                dcc.Graph(
                    figure=px.pie(df, values='count_liquidation', names='token_borrow_mode')
                ),
                dcc.Graph(
                    figure=px.bar(df, x='token_borrow_mode', y='net_outstanding', color='liquidated')
                )
            ], style={'display': 'flex'})
        ]),
        dcc.Tab(label='Liquidations', children=[
            html.Div([
                dcc.Graph(
                    figure=px.scatter(df, x='total_borrow', y='total_liquidation', color='liquidated',
                                      hover_data=['address', 'avg_liquidation_amount', 'avg_liquidation_to_avg_borrow'])
                ),
                dcc.Graph(
                    figure=px.histogram(df, x='std_liquidation_amount', color='liquidated')
                )
            ], style={'display': 'flex'})
        ]),
        dcc.Tab(label='Transactions', children=[
            html.Div([
                dcc.Graph(
                    figure=px.scatter(df, x='total_deposit', y='total_redeem', color='token_borrow_mode',
                                      hover_data=['address', 'avg_deposit_amount', 'avg_redeem_amount'])
                ),
                dcc.Graph(
                    figure=px.bar(df, x='token_borrow_mode', y='net_deposits', color='liquidated')
                )
            ], style={'display': 'flex'})
        ])
    ])
])


if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Graph(id='liquidation-stats'),
    dcc.RangeSlider(
        id='liquidation-range-slider',
        min=df['avg_liquidation_amount'].min(),
        max=df['avg_liquidation_amount'].max(),
        value=[df['avg_liquidation_amount'].min(), df['avg_liquidation_amount'].max()],
        marks={
            int(value): str(value) for value in range(int(df['avg_liquidation_amount'].min()), int(df['avg_liquidation_amount'].max()) + 1, 10)
        }
    )
])

@app.callback(
    Output('liquidation-stats', 'figure'),
    [Input('liquidation-range-slider', 'value')])
def update_graph(selected_range):
    filtered_df = df[(df['avg_liquidation_amount'] >= selected_range[0]) & (df['avg_liquidation_amount'] <= selected_range[1])]
    trace1 = go.Histogram(
        x=filtered_df['avg_liquidation_amount'],
        name='Average Liquidation Amount'
    )
    trace2 = go.Histogram(
        x=filtered_df['count_liquidation'],
        name='Count of Liquidations'
    )
    layout = go.Layout(
        title='Liquidation Statistics',
        xaxis=dict(title='Value'),
        yaxis=dict(title='Count'),
        bargap=0.1,
        barmode='overlay'
    )
    return {'data': [trace1, trace2], 'layout': layout}

if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Graph(id='repay-borrow-ratio'),
    dcc.Slider(
        id='ratio-slider',
        min=df['count_repays_to_count_borrows'].min(),
        max=df['count_repays_to_count_borrows'].max(),
        value=df['count_repays_to_count_borrows'].mean(),
        marks={
            int(value): str(value) for value in range(int(df['count_repays_to_count_borrows'].min()), int(df['count_repays_to_count_borrows'].max()) + 1, 10)
        }
    )
])

@app.callback(
    Output('repay-borrow-ratio', 'figure'),
    [Input('ratio-slider', 'value')])
def update_graph(selected_ratio):
    filtered_df = df[df['count_repays_to_count_borrows'] >= selected_ratio]
    trace = go.Histogram(
        x=filtered_df['count_repays_to_count_borrows'],
        name='Count of Repays to Count of Borrows'
    )
    layout = go.Layout(
        title='Ratio of Repays to Borrows',
        xaxis=dict(title='Ratio'),
        yaxis=dict(title='Count')
    )
    return {'data': [trace], 'layout': layout}

if __name__ == '__main__':
    app.run_server(debug=True)
```
